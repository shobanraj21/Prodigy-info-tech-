# Let's assume you want to visualize the distribution of population across different countries

## Drop rows with missing values in 'Country Name' and '2022' columns

```python
data.dropna(subset=['Country Name', '2022'], inplace=True)
```

 ## Extract relevant columns

```python

countries = data['Country Name']
population = data['2022']
```

## Plotting bar chart

```python

plt.figure(figsize=(15, 8))
plt.bar(countries[:10], population[:10], color='skyblue')
plt.title('Top 10 Countries by Population in 2022')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Drop rows with missing values in '2022' column

```python
data.dropna(subset=['2022'], inplace=True)
```

## Extract population data

```python

population = data['2022']
```

## Plot histogram

```python

plt.figure(figsize=(10, 6))
plt.hist(population, bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Population in 2022')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
